```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
game_data = pd.read_csv("backloggd_games.csv" )
game_data_clean = game_data[['Title', 'Genres']]
game_data_clean = game_data_clean.rename_axis('Id')
```

```python
game_name_map = {}
game_genre_map = {}

name_genre_map = {}


for id, game in game_data_clean.iterrows():
    
    game_genre_map[id] = game['Genres']
    game_name_map[id] = game["Title"]

    name_genre_map[id] = ({
        "Title": game["Title"],
        "Genres": game["Genres"]
    })


# game_genre_map
# game_name_map
name_genre_map
```

```python
target_game_id = 996
# target_game_id = 4
# target_game_id = 0

target_game_object = name_genre_map[target_game_id]
```

```python
game_similarities = []
for game_id, game_object in name_genre_map.items():
    # print(game_object)
    if game_id == target_game_id:
        continue

    games_intersect = set(game_object['Genres']).intersection(set(target_game_object['Genres']))
    games_union = set(game_object['Genres']).union(set(target_game_object['Genres']))

    jaccard = len(games_intersect) / len(games_union)

    game_similarities.append({
        "game": game_id,
        "jaccard": jaccard
    })
```

```python
similarity_df = pd.DataFrame(game_similarities, columns=["game", "jaccard"])
similarity_df
```

```python
most_similar = similarity_df.sort_values(by="jaccard", ascending=False).head(10)

for this_game in most_similar["game"]:
    print(this_game, name_genre_map[this_game]["Title"])
    print("\t", name_genre_map[this_game]["Genres"])
```
